```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Supongamos que ya tienes un DataFrame llamado tabla2
# tabla2 = pd.read_csv('ruta/a/tu/archivo.csv')  # Si necesitas cargar los datos desde un archivo

# Seleccionar solo las columnas numéricas
tabla1_numerica = tabla1.select_dtypes(include=['number'])

# Obtener la matriz de correlación
matriz_correlacion = tabla1_numerica.corr()

# Configurar el estilo de fondo oscuro
plt.style.use('dark_background')

# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 8))

# Dibujar el mapa de calor con la paleta husl
sns.heatmap(matriz_correlacion, annot=True, cmap='cubehelix', fmt='.2f')

# Mostrar el gráfico
plt.show()
```

```python
# Agrupar por Study_Hours_per_Week y calcular la media de GPA
df_grouped = tabla1.groupby("Study_Hours_per_Week")["GPA"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(data=df_grouped, x="Study_Hours_per_Week", y="GPA", marker="o")

plt.xlabel("Study Hours per Week")
plt.ylabel("Average GPA")
plt.title("Average GPA vs Study Hours per Week")
plt.grid(True)
plt.show()
```

```python
# Crear una nueva columna con los rangos de horas de estudio (de 5 en 5)
tabla1["Study_Hours_Binned"] = (tabla1["Study_Hours_per_Week"] // 5) * 5

# Agrupar por los rangos y calcular la media de GPA
df_grouped = tabla1.groupby("Study_Hours_Binned")["GPA"].mean().reset_index()

# Gráfico de líneas
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_grouped, x="Study_Hours_Binned", y="GPA", marker="o")

plt.xlabel("Study Hours per Week (Binned)")
plt.ylabel("Average GPA")
plt.title("Average GPA vs Study Hours per Week (Binned)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer fondo oscuro
plt.style.use("dark_background")

# Crear una nueva columna con los rangos de horas de estudio (de 5 en 5)
tabla1["Study_Hours_Binned"] = (tabla1["Study_Hours_per_Week"] // 5) * 5

# Agrupar por los rangos y calcular la media de GPA
df_grouped = tabla1.groupby("Study_Hours_Binned")["GPA"].mean().reset_index()

# Crear la figura
plt.figure(figsize=(10, 6))

# Gráfico de líneas con la paleta "husl"
sns.lineplot(data=df_grouped, x="Study_Hours_Binned", y="GPA", marker="o", color=sns.color_palette("husl", 1)[0])

# Etiquetas y título
plt.xlabel("Study Hours per Week (Binned)", fontsize=12, color="white")
plt.ylabel("Average GPA", fontsize=12, color="white")
plt.title("Average GPA vs Study Hours per Week (Binned)", fontsize=14, color="white")

# Ajustes estéticos
plt.grid(True, linestyle="--", alpha=0.5)  # Cuadrícula tenue para fondo oscuro

# Mostrar gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer fondo oscuro
plt.style.use("dark_background")

# Crear una nueva columna con los rangos de horas de estudio (de 5 en 5)
tabla1["Study_Hours_Binned"] = (tabla1["Study_Hours_per_Week"] // 5) * 5

# Agrupar por los rangos y calcular la media de GPA
df_grouped = tabla1.groupby("Study_Hours_Binned")["GPA"].max().reset_index()

# Crear la figura
plt.figure(figsize=(10, 6))

# Gráfico de líneas con la paleta "husl"
sns.lineplot(data=df_grouped, x="Study_Hours_Binned", y="GPA", marker="o", color=sns.color_palette("husl", 1)[0])

# Etiquetas y título
plt.xlabel("Study Hours per Week (Binned)", fontsize=12, color="white")
plt.ylabel("Max GPA", fontsize=12, color="white")
plt.title("Max GPA vs Study Hours per Week (Binned)", fontsize=14, color="white")

# Ajustes estéticos
plt.grid(True, linestyle="--", alpha=0.5)  # Cuadrícula tenue para fondo oscuro

# Mostrar gráfico
plt.show()
```

```python
# Calcular la media, el máximo y el mínimo de GPA para cada grupo de horas de estudio
gpa_stats_by_study_hours = tabla1.groupby('Study_Hours_per_Week')['GPA'].agg(['mean', 'max', 'min'])
# Mostrar el resultado
print(gpa_stats_by_study_hours)
# Encontrar los valores máximos y mínimos de GPA junto con sus horas de estudio
max_gpa = tabla1.loc[tabla1['GPA'].idxmax()]
min_gpa = tabla1.loc[tabla1['GPA'].idxmin()]
print("\nMáximo GPA:")
print(max_gpa)
print("\nMínimo GPA:")
print(min_gpa)
```
